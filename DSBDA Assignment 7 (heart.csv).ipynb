```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
ds_heart=pd.read_csv("Heart.csv")
```

```python
ds_heart
```

```python
ds_heart.info()
```

```python
import seaborn as sns
%matplotlib notebook
```

```python
plt.scatter(x='age',y='restecg',data=ds_heart,c='g')
plt.scatter(x='age',y='chol',data=ds_heart, c='r',marker='+')
plt.ylabel('age')
plt.ylabel('chol and restecg')
plt.legend(title='scatter_plot')
plt.show()
```

```python
plt.figure()
sns.lmplot(x='age' , y='chol', data=ds_heart, fit_reg=False)
```

```python
plt.figure()
ds_heart.plot(kind='hist')
```

```python
plt.figure()
sns.displot(ds_heart.age)
```

```python
plt.figure()
plt.xlabel('cp')
plt.ylabel('count')
plt.legend()
ds_heart.cp.value_counts().plot(kind='bar')
```

```python
plt.figure()
sns.countplot(ds_heart.cp)
```

```python
plt.figure()
sns.lineplot(x='age', y='chol', data=ds_heart)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

ds_heart['cp'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)

plt.title('Distribution of Chest Pain Types')

plt.ylabel('')

plt.show()
```
