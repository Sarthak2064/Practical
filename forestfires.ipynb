```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("forestfires.csv") 
```

```python
df.head()
```

```python
df.info()
```

```python
#scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temp', y='wind', data=df)
plt.title('temp vs wind')
plt.xlabel('temp')
plt.ylabel('wind')
plt.show()
```

```python
#histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['area'], bins=30, kde=True)
plt.title('Distribution of Forest Fire Area')
plt.xlabel('area')
plt.ylabel('wind')
plt.show()

```

```python
#pie chart
plt.figure(figsize=(8, 8))
df['month'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Forest Fires by month')
plt.ylabel('')
plt.show()

```

```python
#bar graph
plt.figure(figsize=(10, 6))
sns.countplot(x='month', data=df)
plt.title('Frequency of Forest Fires by Month')
plt.xlabel('month')
plt.ylabel('Frequency')
plt.show()
```

```python
#line graph
plt.figure(figsize=(10, 6))
sns.lineplot(x='day', y='temp', data=df, estimator='mean', ci=None)
plt.title('Average Temperature by Month')
plt.xlabel('day')
plt.ylabel('Temperature (Celsius)')
plt.show()

```
