```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
ds_airquality=pd.read_csv("AirQuality.csv")
```

```python
ds_airquality.info()
```

```python
import seaborn as sns
%matplotlib notebook
```

```python
plt.scatter(x='Ozone',y='Solar.R',data=ds_airquality,c='g')
plt.scatter(x='Ozone',y='Solar.R',data=ds_airquality, c='r',marker='+')
plt.ylabel('Ozone')
plt.ylabel('Ozone and Solar.R')
plt.legend(title='scatter_plot')
plt.show()
```

```python
plt.figure()
sns.lmplot(x='Ozone' , y='Solar.R', data=ds_airquality, fit_reg=False)
```

```python
plt.figure()
ds_airquality.plot(kind='hist')
```

```python
plt.figure()
sns.displot(ds_airquality.Wind)
```

```python
plt.figure()
plt.xlabel('Month')
plt.ylabel('count')
plt.legend()
ds_airquality.Month.value_counts().plot(kind='bar')
```

```python
plt.figure()
sns.countplot(ds_airquality.Month)
```

```python
plt.figure()
sns.lineplot(x='Temp', y='Month', data=ds_airquality)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

ds_airquality['Month'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140)

plt.title('Distribution of Months')

plt.ylabel('')

plt.show()
```
